```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
print(df)
```

```python
#line chart(trend)
plt.plot(df['Days'], df['Study_Hours'], marker='o', linestyle='-', color='r', label='Study Hours')
plt.title('Swastik')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
plt
```

```python
#Advance line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Days'], df['Study_Hours'],
         marker='o', linestyle='-', color='r', label='Study Hours')
plt.plot(df['Days'],df['Marks'],
         marker='o', linestyle='-', color='b', label='Marks')
plt.title('Swastik')
plt.xlabel('Days')
plt.ylabel('Study Hours/Marks')
plt.legend()
plt.show()

```

```python
#2 Bar chart (Comparison)
plt.bar(df['Days'], df['Marks'], color='skyblue')
plt.title('Swastik')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
#Advanced BAr graph
bars = plt.bar(df['Days'], df['Marks'], color='skyblue')
for bar in bars:
     y = bar.get_height()
     plt.text(bar.get_x() + bar.get_width() / 2, y, y, ha='center', va='bottom') # Added ha and va for better positioning
plt.title('Swastik')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y') #add horizontal grid for readability
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(len(df['Days']))
width = 0.35
plt.bar(x-width/2,df['Study_Hours'],width,label='Study Hours')
plt.bar(x+width/2,df['Marks'],width,label='Marks')
plt.xticks(x,df['Days'])
plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Swastik')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (more values for better histogram)

marks = [50, 60, 55, 70, 65, 75, 80, 85, 90, 45, 68, 72, 77, 82, 88]

# Calculate mean

mean_value = np.mean(marks)

plt.figure(figsize=(7,4))

plt.hist(marks, bins=6, color= 'skyblue')

# Mean line

plt.axvline(mean_value,color='red', linestyle='--', linewidth=2, label='Mean')

plt.title("Swastik")

plt.xlabel("Marks")

plt.ylabel("Frequency")

plt.legend()

plt.grid(axis='y')

plt.show()
```

```python
 #4. Scatter Plot (Relationship)
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)

plt.scatter(df['Study_Hours'], df['Marks'])
plt.title("SWASTIK")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
#conditional scatter plot
#Conditional scatter plot

#Example category

df['Result']=['Fail', 'Pass', 'Fail', 'Pass', 'Pass']

colors=['red' if r=='Fail' else 'green' for r in df['Result']]

plt.scatter(df['Study_Hours'], df['Marks'], c=colors)

plt.title("SWASTIK")

plt.xlabel("Study Hours")

plt.ylabel("Marks")

plt.show()
```
